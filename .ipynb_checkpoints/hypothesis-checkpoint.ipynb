# Challenge 2 - Hypothesis Tests

#### In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
#Hypothesis:
#H0 is that the hourly salary is equal to 30$
#H1 is that the hourly salary is significantly different from 30$ 
```

```python
data['hourly_rate'].isna().sum()
```

```python
#I drop the nan values here 
```

```python
data_drop = data['hourly_rate'].dropna()
```

```python
data_drop.isna().sum()
```

```python
# Your code here:
import scipy.stats 
from scipy import stats
hourly_rate = data_drop
confidence_level = 0.95
degrees_freedom = len(hourly_rate) - 1 
sample_mean = np.mean(hourly_rate)
sample_standard_error = scipy.stats.sem(hourly_rate) 
confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

```python
from scipy.stats import ttest_1samp
stat, pval = ttest_1samp(data_drop, 30)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

```python
#The 30$/hr salary is outside our confidence interval from more than 10 % with 95 % certainty. As we can observe in stat. 
```

```python
#From the analysis I can conclude that we don't have enough evidence to support H0
#therefore we reject the Null Hypothesis because the pvalue is less than 0.05, which was our significance value. 
#So yes, the hourly wage of all hourly workers is significantly different from $30/hr.
```

# Challenge 3: To practice - Constructing Confidence Intervals

#### While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 


#### To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


#### In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

#### The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

#### To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:
hourly_rate = data_drop
confidence_level = 0.95
degrees_freedom = len(hourly_rate) - 1 
sample_mean = np.mean(hourly_rate)
sample_standard_error = scipy.stats.sem(hourly_rate) 
confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

# Bonus Challenge - Hypothesis Tests of Proportions

#### Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

#### You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

#### In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
#Hypothesis:
#H0 is that the number of hourly workers is equal to 25 % of the total workers. 
#H1 is that the number of hourly workers is NOT equal to 25 % of the total workers. 
```

```python
# Your code here:
from statsmodels.stats.proportion import proportions_ztest
total = len(data)
hourly_workers = len(data_drop)
proportions_ztest(hourly_workers, total, value=.25)
```

```python
#We can reject the Null Hypothesis H0 which states that the porportion of workers is equal to 25 % . pvalue less than 0.05. 
#It may be less due to the stat value, but Im not sure of this point. 
```
